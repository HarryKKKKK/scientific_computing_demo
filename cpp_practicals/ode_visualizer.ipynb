```python
import matplotlib.pyplot as plt
import numpy as np

def read_data(filename):
    t, y = np.loadtxt(filename, unpack=True)
    return t, y

t_euler, y_euler = read_data("euler_res.dat")
t_rk, y_rk = read_data("runge_kutta.dat")
t_analytic, y_analytic = read_data("analytic_res.dat")

plt.figure(figsize=(8, 5))
plt.plot(t_analytic, y_analytic, 'k-', lw=2, label='Analytic solution')
plt.plot(t_euler, y_euler, 'r--', lw=1.5, label='Euler method')
plt.plot(t_rk, y_rk, 'b-.', lw=1.5, label='Runge-Kutta (2nd order)')

plt.title(r"Solution of $\frac{dy}{dt} = \sqrt{y}$, $y(0)=1$")
plt.xlabel("t")
plt.ylabel("y(t)")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
